## Visualising Dijkstra for Fat tree and Jellyfish

* We compute a 686 version of Fat tree and Jelly Fish as per the paper

```python
ftt_14 = FatTreeTopo(14)
n_ftt_14 = NetworkxVisualizationComposer()
ftt_14.add_composable(n_ftt_14)

ftt_14.generate_fat_tree_structure()

ftt_14.num_hosts
```

```python
jf_14 = JellyFishTopo(14)
n_jf_14 = NetworkxVisualizationComposer()
jf_14.add_composable(n_jf_14)

jf_14.generate_jellyfish_structure()

jf_14.num_hosts
```

```python
all_dj_ftt_14 = ftt_14.compute_dijikstra_for_all_hosts()
dj_ctr_all_dj_ftt_14 = Counter(all_dj_ftt_14)
```

```python
all_dj_jf_14 = jf_14.compute_dijikstra_for_all_hosts()
dj_ctr_all_dj_jf_14 = Counter(all_dj_jf_14)
```

```python
pd_jf_ftt_merged = pd.concat(
    [
    pd.DataFrame.from_dict(dj_ctr_all_dj_ftt_14, orient='index', columns=['ftt_path_length']),
    pd.DataFrame.from_dict(dj_ctr_all_dj_jf_14, orient='index', columns=['jf_path_length'])
    ], 
    axis=1
)
```

```python
pd_jf_ftt_merged = pd_jf_ftt_merged.fillna(0)
```

```python
normalized_pf_jf_ftt = pd_jf_ftt_merged.div(pd_jf_ftt_merged.max().max(), axis=0).sort_index()
```

```python
normalized_pf_jf_ftt
```

```python
normalized_pf_jf_ftt.plot( kind="bar")
```

```python
ftt_4_centrality = {
    "8_shortest": ftt_4.calculate_centrality(ftt_4_yen[1]),
    "8_ecmp": ftt_4.calculate_centrality(ftt_4_yen[2]),
    "64_ecmp": ftt_4.calculate_centrality(ftt_4_yen[3])
}

jf_4_centrality = {
    "8_shortest": jf_4.calculate_centrality(jf_4_yen[1]),
    "8_ecmp": jf_4.calculate_centrality(jf_4_yen[2]),
    "64_ecmp": jf_4.calculate_centrality(jf_4_yen[3])
}

dc_4_centrality = {
    "8_shortest": dc_4_1.calculate_centrality(dc_4_yen[1]),
    "8_ecmp": dc_4_1.calculate_centrality(dc_4_yen[2]),
    "64_ecmp": dc_4_1.calculate_centrality(dc_4_yen[3])
}

bc_4_centrality = {
    "8_shortest": bc_4_1.calculate_centrality(bc_4_yen[1]),
    "8_ecmp": bc_4_1.calculate_centrality(bc_4_yen[2]),
    "64_ecmp": bc_4_1.calculate_centrality(bc_4_yen[3])
}
```

```python
def inverse_centrality(centrality_map):
    temp_dict = {}
    for n, o in centrality_map.items():
        res = {}
        for k,v in o.items():
            res[v] = (res.get(v) or 0) + 1
        temp_dict[n] = res
    return temp_dict
```

```python
pd.DataFrame(inverse_centrality(ftt_4_centrality)).fillna(0).sort_index().plot(title="ftt-4-ecmp")
```

```python
pd.DataFrame(inverse_centrality(jf_4_centrality)).fillna(0).sort_index().plot(title="jf-4-ecmp")
```

```python
pd.DataFrame(inverse_centrality(dc_4_centrality)).fillna(0).sort_index().plot(title="dc-4-ecmp")
```

```python
pd.DataFrame(inverse_centrality(bc_4_centrality)).fillna(0).sort_index().plot(title="bc-4-ecmp")
```
